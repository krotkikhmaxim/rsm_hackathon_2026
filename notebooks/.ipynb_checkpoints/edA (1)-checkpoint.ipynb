## 3. Подготовка данных

Что делаем:
- приводим даты;
- нормализуем бинарные и числовые поля;
- извлекаем временные признаки;
- создаем категории масштаба инфраструктуры;
- объединяем incidents со справочником угроз.

```python
thr = thr.copy()
inc = inc.copy()

# Даты в справочнике угроз
for col in ['Дата включения угрозы в БнД УБИ', 'Дата последнего изменения данных']:
    if col in thr.columns:
        thr[col] = pd.to_datetime(thr[col], dayfirst=True, errors='coerce')

# Даты в инцидентах
for col in ['Дата инцидента', 'Региональное время']:
    if col in inc.columns:
        inc[col] = pd.to_datetime(inc[col], dayfirst=True, errors='coerce')

# Числовые поля
for col in ['Количество хостов', 'Успех', 'Код реализованной угрозы', 'Код предприятия']:
    if col in inc.columns:
        inc[col] = pd.to_numeric(inc[col], errors='coerce')

if 'Идентификатор УБИ' in thr.columns:
    thr['Идентификатор УБИ'] = pd.to_numeric(thr['Идентификатор УБИ'], errors='coerce')

# Временные признаки incidents
inc['Год_инцидента'] = inc['Дата инцидента'].dt.year
inc['Месяц_инцидента'] = inc['Дата инцидента'].dt.month
inc['Квартал_инцидента'] = inc['Дата инцидента'].dt.quarter
inc['День_недели_инцидента'] = inc['Дата инцидента'].dt.dayofweek
inc['Час_инцидента'] = inc['Региональное время'].dt.hour
inc['День_месяца_инцидента'] = inc['Дата инцидента'].dt.day
inc['Выходной'] = inc['День_недели_инцидента'].isin([5, 6]).astype('Int64')

season_map = {
    12: 'Зима', 1: 'Зима', 2: 'Зима',
    3: 'Весна', 4: 'Весна', 5: 'Весна',
    6: 'Лето', 7: 'Лето', 8: 'Лето',
    9: 'Осень', 10: 'Осень', 11: 'Осень'
}
inc['Сезон'] = inc['Месяц_инцидента'].map(season_map)

# Категоризация инфраструктуры
if 'Количество хостов' in inc.columns:
    inc['Размер инфраструктуры'] = pd.cut(
        inc['Количество хостов'],
        bins=[-np.inf, 250, 1000, 2000, np.inf],
        labels=['Малая', 'Средняя', 'Крупная', 'Очень крупная']
    )

# Признаки по справочнику угроз
if 'Дата включения угрозы в БнД УБИ' in thr.columns:
    thr['Год_включения_угрозы'] = thr['Дата включения угрозы в БнД УБИ'].dt.year
    thr['Месяц_включения_угрозы'] = thr['Дата включения угрозы в БнД УБИ'].dt.month

if {'Дата включения угрозы в БнД УБИ', 'Дата последнего изменения данных'}.issubset(thr.columns):
    thr['Дней_в_базе'] = (
        thr['Дата последнего изменения данных'] - thr['Дата включения угрозы в БнД УБИ']
    ).dt.days

# Merge
df = inc.merge(
    thr,
    how='left',
    left_on='Код реализованной угрозы',
    right_on='Идентификатор УБИ'
)

print('merged shape:', df.shape)
display(df.head())
```

```python
baseline_df = df.copy()

print('Размер baseline_df:', baseline_df.shape)
display(baseline_df.head())
```

```python
def classify_period(day_of_week, hour):
    """
    day_of_week:
        0 = Пн
        1 = Вт
        2 = Ср
        3 = Чт
        4 = Пт
        5 = Сб
        6 = Вс

    hour:
        час от 0 до 23
    """
    if pd.isna(day_of_week) or pd.isna(hour):
        return np.nan

    day_of_week = int(day_of_week)
    hour = int(hour)

    # Будни
    if day_of_week in [0, 1, 2, 3, 4]:
        if 6 <= hour < 12:
            return 'Будни_Утро'
        elif 12 <= hour < 18:
            return 'Будни_День'
        elif 18 <= hour < 24:
            return 'Будни_Вечер'
        else:
            return 'Будни_Ночь'

    # Выходные
    elif day_of_week in [5, 6]:
        if 6 <= hour < 22:
            return 'Выходной_День'
        else:
            return 'Выходной_Ночь'

    return np.nan
```

```python
baseline_df['Период'] = baseline_df.apply(
    lambda row: classify_period(row['День_недели_инцидента'], row['Час_инцидента']),
    axis=1
)

print('Распределение по таргету "Период":')
display(baseline_df['Период'].value_counts(dropna=False))

print('Доли классов "Период":')
display(baseline_df['Период'].value_counts(normalize=True, dropna=False).round(4))
```

```python
def extract_datetime_features(df, datetime_col):
    """Извлекает признаки из datetime колонки"""
    df = df.copy()
    dt = pd.to_datetime(df[datetime_col])
    
    df['year'] = dt.dt.year
    df['month'] = dt.dt.month
    df['day'] = dt.dt.day
    df['hour'] = dt.dt.hour
    df['dayofweek'] = dt.dt.dayofweek
    df['is_weekend'] = (dt.dt.dayofweek >= 5).astype(int)
    df['quarter'] = dt.dt.quarter
    
    df = df.drop(columns=[datetime_col])
    
    return df
```

```python
def encode_categorical_features(
    df: pd.DataFrame,
    exclude_cols: Optional[List[str]] = None,
    drop_datetime: bool = True,
    fill_na_strategy: str = 'unknown',
    verbose: bool = True
) -> Tuple[pd.DataFrame, Dict[str, LabelEncoder]]:
    """
    Кодирует категориальные признаки в числовые для использования в ML моделях
    
    Parameters
    ----------
    df : pd.DataFrame
        Исходный DataFrame
    exclude_cols : List[str], optional
        Список колонок для исключения (целевая переменная, ID, текст и т.д.)
    drop_datetime : bool, default=True
        Удалять ли datetime колонки
    fill_na_strategy : str, default='unknown'
        Стратегия заполнения пропусков: 'unknown', 'mode', 'zero'
    verbose : bool, default=True
        Выводить ли информацию о процессе
    
    Returns
    -------
    df_encoded : pd.DataFrame
        DataFrame с закодированными признаками
    encoders : Dict[str, LabelEncoder]
        Словарь с LabelEncoder для каждой категориальной колонки
    """
    
    df = df.copy()
    encoders = {}
    
    if verbose:
        print(f"   Исходный размер: {df.shape[0]} строк × {df.shape[1]} колонок")
    
    # ========================================================================
    # Шаг 1: Исключение указанных колонок
    # ========================================================================
    if exclude_cols:
        cols_to_drop = [c for c in exclude_cols if c in df.columns]
        if cols_to_drop and verbose:
            print(f"\n🗑️  Исключено колонок: {len(cols_to_drop)}")
            for col in cols_to_drop:
                print(f"      - {col}")
        df = df.drop(columns=cols_to_drop)
    
    # ========================================================================
    # Шаг 2: Удаление datetime колонок
    # ========================================================================
    if drop_datetime:
        datetime_cols = df.select_dtypes(include=['datetime64[ns]']).columns
        if len(datetime_cols) > 0:
            if verbose:
                print(f"\n📅 Удалено datetime колонок: {len(datetime_cols)}")
                for col in datetime_cols:
                    print(f"      - {col}")
            df = df.drop(columns=datetime_cols)
    
    # ========================================================================
    # Шаг 3: Кодирование категориальных признаков
    # ========================================================================
    cat_cols = df.select_dtypes(include=['object', 'category']).columns
    
    if verbose:
        print(f"\n🏷️  Категориальных колонок для кодирования: {len(cat_cols)}")
    
    for col in cat_cols:
        le = LabelEncoder()
        
        # Заполнение пропусков
        if fill_na_strategy == 'unknown':
            df[col] = df[col].astype(str).fillna('Unknown')
        elif fill_na_strategy == 'mode':
            mode_val = df[col].mode()[0] if len(df[col].mode()) > 0 else 'Unknown'
            df[col] = df[col].astype(str).fillna(mode_val)
        elif fill_na_strategy == 'zero':
            df[col] = df[col].astype(str).fillna('0')
        else:
            df[col] = df[col].astype(str).fillna('Unknown')
        
        # Кодирование
        df[col] = le.fit_transform(df[col])
        encoders[col] = le
        
        if verbose:
            unique_count = len(le.classes_)
            print(f"      ✓ {col}: {unique_count} уникальных значений")
    
    # ========================================================================
    # Шаг 4: Преобразование всех колонок в numeric
    # ========================================================================
    if verbose:
        print(f"\n🔢 Преобразование в numeric...")
    
    for col in df.columns:
        if df[col].dtype not in ['int64', 'float64', 'int32', 'float32']:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
    
    # ========================================================================
    # Шаг 5: Итоговая информация
    # ========================================================================
    if verbose:
        print(f"\n✅ Кодирование завершено!")
        print(f"   Итоговый размер: {df.shape[0]} строк × {df.shape[1]} колонок")
        print(f"   Типы данных:")
        for dtype, count in df.dtypes.value_counts().items():
            print(f"      {dtype}: {count} колонок")
        print("="*70)
    
    return df, encoders

def encode_target_variable(
    df: pd.DataFrame,
    target_col: str,
    verbose: bool = True
) -> Tuple[np.ndarray, np.ndarray, LabelEncoder]:
    """
    Кодирует целевую переменную для классификации
    
    Parameters
    ----------
    df : pd.DataFrame
        Исходный DataFrame
    target_col : str
        Название колонки с целевой переменной
    verbose : bool, default=True
        Выводить ли информацию
    
    Returns
    -------
    y_encoded : np.ndarray
        Закодированная целевая переменная
    classes : np.ndarray
        Уникальные классы (оригинальные названия)
    le : LabelEncoder
        LabelEncoder для обратного преобразования
    """
    
    le = LabelEncoder()
    y_encoded = le.fit_transform(df[target_col].astype(str))
    classes = le.classes_
    
    if verbose:
        print(f"\n🎯 Целевая переменная: '{target_col}'")
        print(f"   Уникальных классов: {len(classes)}")
        print(f"   Распределение:")
        for cls, count in zip(classes, np.bincount(y_encoded)):
            print(f"      {cls}: {count} ({count/len(y_encoded)*100:.1f}%)")
        
        min_count = np.min(np.bincount(y_encoded))
        can_stratify = min_count >= 2
        print(f"   {'✅' if can_stratify else '⚠️'} Stratify: {'Возможно' if can_stratify else 'Невозможно'} (min={min_count})")
    
    return y_encoded, classes, le
```

```python
def prepare_data_dict(df, test_size=0.2, random_state=42):
    """
    Подготовка данных для 4 baseline моделей
    Использует отдельные функции кодирования
    """
    
    df = df.copy()
    data_dict = {}
    
    # МОДЕЛЬ 1: Тип угрозы
    
    exclude_cols_m1 = [
        'Код реализованной угрозы', 'Дата инцидента', 'Региональное время',
        'Дата включения угрозы в БнД УБИ', 'Дата последнего изменения данных',
        'Успех', 'Объект воздействия', 'Замечания', 'Описание', 'Наименование УБИ'
    ]

    df_m1 = extract_datetime_features(df.copy(), 'Дата инцидента')
    df_m1 = extract_datetime_features(df_m1, 'Региональное время')
    
    X1, encoders_1 = encode_categorical_features(
        df=df_m1, exclude_cols=exclude_cols_m1, verbose=False
    )
    
    y1_encoded, classes_1, le_target_1 = encode_target_variable(
        df=df_m1, target_col='Код реализованной угрозы', verbose=False
    )
    
    can_stratify_1 = len(np.unique(y1_encoded)) > 1 and np.min(np.bincount(y1_encoded)) >= 2
    
    X1_train, X1_test, y1_train, y1_test = train_test_split(
        X1, y1_encoded, test_size=test_size, random_state=random_state,
        stratify=y1_encoded if can_stratify_1 else None
    )
    
    data_dict['model1'] = {
        'X_train': X1_train, 'y_train': y1_train,
        'X_test': X1_test, 'y_test': y1_test,
        'classes': classes_1,
        '_label_encoder': le_target_1,
        '_feature_encoders': encoders_1
    }
    
    # МОДЕЛЬ 2: Объект воздействия

    exclude_cols_m2 = [
        'Объект воздействия', 'Дата инцидента', 'Региональное время',
        'Дата включения угрозы в БнД УБИ', 'Дата последнего изменения данных',
        'Успех', 'Код реализованной угрозы', 'Замечания', 'Описание', 'Наименование УБИ'
    ]

    df_m2 = extract_datetime_features(df.copy(), 'Дата инцидента')
    df_m2 = extract_datetime_features(df_m2, 'Региональное время')
    
    X2, encoders_2 = encode_categorical_features(
        df=df_m2, exclude_cols=exclude_cols_m2, verbose=False
    )
    
    y2_encoded, classes_2, le_target_2 = encode_target_variable(
        df=df_m2, target_col='Объект воздействия', verbose=False
    )
    
    can_stratify_2 = len(np.unique(y2_encoded)) > 1 and np.min(np.bincount(y2_encoded)) >= 2
    
    X2_train, X2_test, y2_train, y2_test = train_test_split(
        X2, y2_encoded, test_size=test_size, random_state=random_state,
        stratify=y2_encoded if can_stratify_2 else None
    )
    
    data_dict['model2'] = {
        'X_train': X2_train, 'y_train': y2_train,
        'X_test': X2_test, 'y_test': y2_test,
        'classes': classes_2,
        '_label_encoder': le_target_2,
        '_feature_encoders': encoders_2
    }
    
    # МОДЕЛЬ 3: Успешность атаки

    exclude_cols_m3 = [
        'Успех', 'Дата инцидента', 'Региональное время',
        'Дата включения угрозы в БнД УБИ', 'Дата последнего изменения данных',
        'Замечания', 'Описание', 'Наименование УБИ'
    ]

    df_m3 = extract_datetime_features(df.copy(), 'Дата инцидента')
    df_m3 = extract_datetime_features(df_m3, 'Региональное время')
    
    X3, encoders_3 = encode_categorical_features(
        df=df_m3, exclude_cols=exclude_cols_m3, verbose=False
    )
    
    y3_encoded, classes_3, le_target_3 = encode_target_variable(
        df=df_m3, target_col='Успех', verbose=False
    )
    
    can_stratify_3 = len(np.unique(y3_encoded)) > 1 and np.min(np.bincount(y3_encoded)) >= 2
    
    X3_train, X3_test, y3_train, y3_test = train_test_split(
        X3, y3_encoded, test_size=test_size, random_state=random_state,
        stratify=y3_encoded if can_stratify_3 else None
    )
    
    data_dict['model3'] = {
        'X_train': X3_train, 'y_train': y3_train,
        'X_test': X3_test, 'y_test': y3_test,
        'classes': classes_3,
        '_label_encoder': le_target_3,
        '_feature_encoders': encoders_3
    }

    # МОДЕЛЬ 4: Период атаки (Временной ряд)
    
    df_m4 = df.copy()

    df_sorted = df_m4.sort_values('Региональное время').reset_index(drop=True)
    
    def get_time_period(hour):
        if 0 <= hour < 6:
            return 'Night (00-06)'
        elif 6 <= hour < 12:
            return 'Morning (06-12)'
        elif 12 <= hour < 18:
            return 'Afternoon (12-18)'
        else:
            return 'Evening (18-24)'
    
    if 'Час_инцидента' in df_sorted.columns:
        periods = [get_time_period(h) for h in df_sorted['Час_инцидента']]
    else:
        periods = [get_time_period(ts.hour) for ts in df_sorted['Региональное время']]
    
    timestamps = df_sorted['Региональное время'].values
    split_idx = int(len(df_sorted) * (1 - test_size))
    
    data_dict['model4'] = {
        'timestamps': timestamps,
        'y': np.array(periods),
        'classes': np.array(['Night (00-06)', 'Morning (06-12)', 'Afternoon (12-18)', 'Evening (18-24)']),
        '_full_df_sorted': df_sorted,
        '_split_idx': split_idx
    }
    
    print(f"   Модель 1: {X1_train.shape[1]} признаков, {len(classes_1)} классов")
    print(f"   Модель 2: {X2_train.shape[1]} признаков, {len(classes_2)} классов")
    print(f"   Модель 3: {X3_train.shape[1]} признаков, {len(classes_3)} классов")
    print(f"   Модель 4: {len(timestamps)} записей (временной ряд)")
    
    return data_dict
```

```python
def vis(X_train, y_train, y_pred, rf, y_test, y_proba_aligned, all_classes=None, plt_matrix=True):
    # Расчёт log_loss с явным указанием классов
    if all_classes:
        log_loss_value = log_loss(y_test, y_proba_aligned, labels=all_classes)
    else:
        log_loss_value = None

    # Cross-validation
    cv_scores = cross_val_score(rf, X_train, y_train, cv=5, scoring='accuracy')
    print(f"\n📈 Cross-Validation Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std()*2:.4f})")

    # Feature Importance
    feature_columns = X_train.columns if hasattr(X_train, 'columns') else [f'feat_{i}' for i in range(X_train.shape[1])]
    feature_importance = pd.DataFrame({
        'feature': feature_columns,
        'importance': rf.feature_importances_
    }).sort_values('importance', ascending=False)

    print("\n🔝 Top-10 Important Features:")
    print(feature_importance.head(10).to_string(index=False))

    # Визуализация
    plt.figure(figsize=(10, 6))
    top_10 = feature_importance.head(10)
    plt.barh(range(len(top_10)), top_10['importance'].values)
    plt.yticks(range(len(top_10)), top_10['feature'].values)
    plt.xlabel('Importance')
    plt.title('Top-10 Feature Importance — Model 1 (Threat Type)')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    rows, cols = np.where(cm > 0)
    values = cm[rows, cols]
    df_scatter = pd.DataFrame({'True': rows, 'Predicted': cols, 'Count': values})
    
    if plt_matrix:
        plt.figure(figsize=(10, 8))
        plt.scatter(df_scatter['True'], df_scatter['Predicted'], s=df_scatter['Count'] * 10, alpha=0.6)
        plt.xlabel('True Label')
        plt.ylabel('Predicted Label')
        plt.title('Error Patterns (Dot Size = Count)')
        plt.grid(True, alpha=0.3)
        plt.show()

    return cm, cv_scores, feature_importance, log_loss_value
```

```python
# МОДЕЛЬ 1: Тип угрозы (Random Forest)

def baseline_model_1(X_train, y_train, X_test, y_test, class_names=None):
    print("="*70)
    print("🔵 МОДЕЛЬ 1 — Тип угрозы")
    print("="*70)
          
    train_classes = set(np.unique(y_train))
    test_classes = set(np.unique(y_test))
    missing_in_test = train_classes - test_classes
    extra_in_test = test_classes - train_classes

    if missing_in_test:
        print(f"[ВНИМАНИЕ] Классы отсутствуют в тестовой выборке: {sorted(missing_in_test)}")
    if extra_in_test:
        print(f"[ВНИМАНИЕ] Новые классы в тестовой выборке: {sorted(extra_in_test)}")

    # Определяем полный набор классов
    all_classes = sorted(train_classes | test_classes)
    n_classes = len(all_classes)

    rf = RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=42,
        n_jobs=-1,
        class_weight='balanced'
    )

    print(f"\n📊 Обучение Random Forest...")
    print(f"   Train samples: {len(y_train)}")
    print(f"   Features: {X_train.shape[1]}")
    print(f"   Classes: {n_classes}")

    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    y_proba = rf.predict_proba(X_test)

    # Выравниваем размерность вероятностей
    if y_proba.shape[1] != n_classes:
        # Создаём массив с правильной размерностью
        aligned_proba = np.zeros((y_proba.shape[0], n_classes))
        # Находим индексы классов, которые есть в предсказаниях
        pred_classes = rf.classes_
        for i, cls in enumerate(pred_classes):
            if cls in all_classes:
                idx = all_classes.index(cls)
                aligned_proba[:, idx] = y_proba[:, i]
        y_proba_aligned = aligned_proba
    else:
        y_proba_aligned = y_proba

    cm, cv_scores, feature_importance, log_loss_value = vis(X_train, y_train, y_pred, rf, y_test, y_proba_aligned, all_classes)

    return {
        'model': rf,
        'metrics': {
            'accuracy': accuracy_score(y_test, y_pred),
            'log_loss': log_loss_value,
            'cv_accuracy_mean': cv_scores.mean(),
            'cv_accuracy_std': cv_scores.std()
        },
        'feature_importance': feature_importance,
        'confusion_matrix': cm,
        'predictions': y_pred,
        'probabilities': y_proba_aligned,
        'X_test': X_test
    }
```

```python
# МОДЕЛЬ 2: Объект воздействия (Random Forest)


def baseline_model_2(X_train, y_train, X_test, y_test, class_names=None):
    
    print("="*70)
    print("🔵 МОДЕЛЬ 2 — Объект воздействия")
    print("="*70)
    
    # Валидация классов
    train_classes = set(np.unique(y_train))
    test_classes = set(np.unique(y_test))
    missing_in_test = train_classes - test_classes
    extra_in_test = test_classes - train_classes

    if missing_in_test:
        print(f"[ВНИМАНИЕ] Классы отсутствуют в тестовой выборке: {sorted(missing_in_test)}")
    if extra_in_test:
        print(f"[ВНИМАНИЕ] Новые классы в тестовой выборке: {sorted(extra_in_test)}")

    all_classes = sorted(train_classes | test_classes)
    n_classes = len(all_classes)

    rf = RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=42,
        n_jobs=-1,
        class_weight='balanced'
    )

    print(f"\n📊 Обучение Random Forest...")
    print(f"   Train samples: {len(y_train)}")
    print(f"   Features: {X_train.shape[1]}")
    print(f"   Classes: {n_classes}")

    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    y_proba = rf.predict_proba(X_test)

    # Выравнивание размерности вероятностей
    if y_proba.shape[1] != n_classes:
        aligned_proba = np.zeros((y_proba.shape[0], n_classes))
        pred_classes = rf.classes_
        for i, cls in enumerate(pred_classes):
            if cls in all_classes:
                idx = all_classes.index(cls)
                aligned_proba[:, idx] = y_proba[:, i]
        y_proba_aligned = aligned_proba
    else:
        y_proba_aligned = y_proba

    cm, cv_scores, feature_importance, log_loss_value = vis(X_train, y_train, y_pred, rf, y_test, y_proba_aligned, all_classes)

    return {
        'model': rf,
        'metrics': {
            'accuracy': accuracy_score(y_test, y_pred),
            'log_loss': log_loss_value,
            'cv_accuracy_mean': cv_scores.mean(),
            'cv_accuracy_std': cv_scores.std()
        },
        'feature_importance': feature_importance,
        'confusion_matrix': cm,
        'predictions': y_pred,
        'probabilities': y_proba_aligned,
        'X_test': X_test
    }
```

```python
# МОДЕЛЬ 3: Успешность атаки (Random Forest, Binary)

def baseline_model_3(X_train, y_train, X_test, y_test, class_names=None):
    """
    Baseline для Модели 3: Успешность атаки
    """
    print("="*70)
    print("🔵 МОДЕЛЬ 3 — Успешность атаки")
    print("="*70)
    
    rf = RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=42,
        n_jobs=-1,
        class_weight='balanced'
    )
    
    print(f"\n📊 Обучение Random Forest...")
    print(f"   Train samples: {len(y_train)}")
    print(f"   Features: {X_train.shape[1]}")
    print(f"   Classes: {np.unique(y_train)}")
    
    rf.fit(X_train, y_train)
    
    y_pred = rf.predict(X_test)
    y_proba = rf.predict_proba(X_test)
    
    cm, cv_scores, feature_importance, _ = vis(X_train, y_train, y_pred, rf, y_test, y_proba, plt_matrix=False)
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1])
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'Random Forest, AUC={roc_auc_score(y_test, y_proba[:, 1]):.3f}')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve — Random Forest')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

    # Confusion Matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', 
                xticklabels=['Failed', 'Success'],
                yticklabels=['Failed', 'Success'])
    plt.title('Confusion Matrix')
    plt.ylabel('True')
    plt.xlabel('Predicted')
    plt.tight_layout()
    plt.show()
    
    return {
        'model': rf,
        'metrics': {
            'accuracy': accuracy_score(y_test, y_pred),
            'roc_auc': roc_auc_score(y_test, y_proba[:, 1]),
            'f1': f1_score(y_test, y_pred),
            'cv_f1_mean': cv_scores.mean(),
            'cv_f1_std': cv_scores.std()
        },
        'feature_importance': feature_importance,
        'confusion_matrix': cm,
        'predictions': y_pred,
        'probabilities': y_proba,
        'X_test': X_test
    }
```

```python
# МОДЕЛЬ 4
class MovingAverageBaseline:
    """
    Baseline на основе скользящего среднего для временных данных
    """
    
    def __init__(self, window_size=20, strategy='mode', alpha=None):
        self.window_size = window_size
        self.strategy = strategy
        self.alpha = alpha
        self.classes_ = None
        self.history = []
        
    def fit(self, X, y):
        self.classes_ = np.unique(y)
        return self
    
    def _get_mode_prediction(self):
        if len(self.history) == 0:
            return np.ones(len(self.classes_)) / len(self.classes_)
        
        window = self.history[-self.window_size:]
        probs = np.array([window.count(cls) / len(window) for cls in self.classes_])
        return probs
    
    def _get_ema_prediction(self):
        alpha = self.alpha or (2 / (self.window_size + 1))
        n_classes = len(self.classes_)
        
        if len(self.history) == 0:
            return np.ones(n_classes) / n_classes
        
        probs = np.zeros(n_classes)
        for class_idx, cls in enumerate(self.classes_):
            hist = np.array([1 if val == cls else 0 for val in self.history])
            ema = hist[0]
            for val in hist[1:]:
                ema = alpha * val + (1 - alpha) * ema
            probs[class_idx] = ema
        
        if probs.sum() > 0:
            probs = probs / probs.sum()
        return probs
    
    def predict_proba(self, X):
        n_samples = len(X) if hasattr(X, '__len__') else 1
        
        if self.strategy == 'ema':
            probs = self._get_ema_prediction()
        else:
            probs = self._get_mode_prediction()
            
        return np.tile(probs, (n_samples, 1))
    
    def predict(self, X):
        proba = self.predict_proba(X)
        return np.argmax(proba, axis=1)
    
    def update_history(self, y_true):
        if hasattr(y_true, 'tolist'):
            self.history.extend(y_true.tolist())
        else:
            self.history.append(y_true)
        if len(self.history) > 2000:
            self.history = self.history[-2000:]


def baseline_model_4(timestamps, y, window_size=20, test_size=0.2, class_names=None):
    """
    Baseline для Модели 4: Период атаки
    """
    print("="*70)
    print("🔵 МОДЕЛЬ 4 — ПЕРИОД АТАКИ (Moving Average Baseline)")
    print("="*70)
    
    # Кодирование
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    class_names_final = class_names if class_names is not None else le.classes_
    
    # Разделение с сохранением временного порядка
    split_idx = int(len(timestamps) * (1 - test_size))
    
    y_train = y_encoded[:split_idx]
    y_test = y_encoded[split_idx:]
    ts_test = timestamps[split_idx:]
    
    print(f"Train: {len(y_train)} | Test: {len(y_test)}")
    print(f"Classes: {class_names_final}")
    
    # Тестирование стратегий
    strategies = [('mode', None), ('ema', 0.1), ('ema', 0.05)]
    results = {}
    
    for strategy, alpha in strategies:
        ma = MovingAverageBaseline(window_size=window_size, strategy=strategy, alpha=alpha)
        ma.fit(None, y_train)
        ma.update_history(y_train)
        
        y_pred = ma.predict(ts_test)
        y_proba = ma.predict_proba(ts_test)
        
        acc = accuracy_score(y_test, y_pred)
        ll = log_loss(y_test, y_proba, labels=ma.classes_)
        
        key = f"{strategy}{f'_α{alpha}' if alpha else ''}"
        results[key] = {'accuracy': acc, 'log_loss': ll, 'model': ma}
        
        print(f"\n📊 Strategy: '{key}' (window={window_size})")
        print(f"   Accuracy: {acc:.4f} | LogLoss: {ll:.4f}")
    
    # Лучшая стратегия
    best_key = max(results, key=lambda x: results[x]['accuracy'])
    best_model = results[best_key]['model']
    
    print(f"\n✅ Лучшая конфигурация: '{best_key}'")
    
    # Финальные предсказания
    y_pred_final = best_model.predict(ts_test)
    y_proba_final = best_model.predict_proba(ts_test)
    
    # Визуализация
    fig, axes = plt.subplots(2, 1, figsize=(14, 8))
    
    ax = axes[0]
    colors = plt.cm.Set2(np.linspace(0, 1, len(class_names_final)))
    for idx, cls in enumerate(class_names_final):
        mask = y_test == idx
        ax.scatter(np.arange(len(ts_test))[mask], [idx] * mask.sum(), 
                   c=[colors[idx]], s=15, label=str(cls), alpha=0.6)
    ax.set_ylabel('Period (encoded)')
    ax.set_title('True Periods over Test Time')
    ax.legend(ncol=2, fontsize=8)
    ax.grid(alpha=0.3)
    
    ax = axes[1]
    ax.plot(np.arange(len(ts_test)), y_test, label='True', alpha=0.7, linewidth=1)
    ax.plot(np.arange(len(ts_test)), y_pred_final, label='Predicted (MA)', 
            alpha=0.7, linewidth=1, linestyle='--')
    ax.set_ylabel('Period (encoded)')
    ax.set_xlabel('Time Index (test set)')
    ax.set_title(f'Moving Average Predictions (window={window_size}, {best_key})')
    ax.legend()
    ax.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Анализ окна
    print("\n🔍 Анализ размера окна:")
    window_results = []
    for ws in [5, 10, 20, 50, 100]:
        ma_test = MovingAverageBaseline(window_size=ws, strategy='mode')
        ma_test.fit(None, y_train)
        ma_test.update_history(y_train)
        y_pred_ws = ma_test.predict(ts_test)
        acc_ws = accuracy_score(y_test, y_pred_ws)
        window_results.append({'window': ws, 'accuracy': acc_ws})
        print(f"   Window {ws:3d} → Accuracy: {acc_ws:.4f}")
    
    # Feature importance (эвристические для MA)
    feature_importance = pd.DataFrame({
        'feature': ['temporal_autocorrelation', 'historical_frequency', 'period_cyclicity'],
        'importance': [0.5, 0.3, 0.2]
    })
    
    print("\n🔝 Top Features (эвристические для MA):")
    print(feature_importance.to_string(index=False))
    
    return {
        'model': best_model,
        'label_encoder': le,
        'strategy': best_key,
        'window_size': window_size,
        'metrics': {
            'accuracy': accuracy_score(y_test, y_pred_final),
            'log_loss': log_loss(y_test, y_proba_final, labels=best_model.classes_)
        },
        'feature_importance': feature_importance,
        'window_analysis': pd.DataFrame(window_results),
        'predictions': y_pred_final,
        'probabilities': y_proba_final,
        'X_test': None
    }
```

```python
def run_all_baselines(data_dict):
    """Запуск всех baseline моделей"""
    results = {}
    
    if 'model1' in data_dict:
        d1 = data_dict['model1']
        results['model1'] = baseline_model_1(
            d1['X_train'], d1['y_train'], d1['X_test'], d1['y_test'],
            class_names=d1.get('classes')
        )
    
    if 'model2' in data_dict:
        d2 = data_dict['model2']
        results['model2'] = baseline_model_2(
            d2['X_train'], d2['y_train'], d2['X_test'], d2['y_test'],
            class_names=d2.get('classes')
        )
    
    if 'model3' in data_dict:
        d3 = data_dict['model3']
        results['model3'] = baseline_model_3(
            d3['X_train'], d3['y_train'], d3['X_test'], d3['y_test'],
            class_names=d3.get('classes')
        )
    
    if 'model4' in data_dict:
        d4 = data_dict['model4']
        results['model4'] = baseline_model_4(
            timestamps=d4['timestamps'],
            y=d4['y'],
            window_size=20,
            class_names=d4.get('classes')
        )
    
    return results
```

```python
def print_summary_table(results):
    """Вывод сводной таблицы всех моделей"""
    print("\n" + "="*70)
    print("📋 СВОДНАЯ ТАБЛИЦА РЕЗУЛЬТАТОВ")
    print("="*70)
    
    summary_data = []
    
    for model_name, result in results.items():
        row = {'Model': model_name}
        
        if model_name == 'model1':
            row['Task'] = 'Multiclass (Threat Type)'
            row['Accuracy'] = f"{result['metrics']['accuracy']:.4f}"
            row['LogLoss'] = f"{result['metrics']['log_loss']:.4f}"
            row['CV Acc'] = f"{result['metrics']['cv_accuracy_mean']:.4f}"
        elif model_name == 'model2':
            row['Task'] = 'Multiclass (Target Object)'
            row['Accuracy'] = f"{result['metrics']['accuracy']:.4f}"
            row['LogLoss'] = f"{result['metrics']['log_loss']:.4f}"
            row['CV Acc'] = f"{result['metrics']['cv_accuracy_mean']:.4f}"
        elif model_name == 'model3':
            row['Task'] = 'Binary (Attack Success)'
            row['Accuracy'] = f"{result['metrics']['accuracy']:.4f}"
            row['ROC-AUC'] = f"{result['metrics']['roc_auc']:.4f}"
            row['F1'] = f"{result['metrics']['f1']:.4f}"
            row['CV F1'] = f"{result['metrics']['cv_f1_mean']:.4f}"
        elif model_name == 'model4':
            row['Task'] = 'Time Series (Attack Period)'
            row['Accuracy'] = f"{result['metrics']['accuracy']:.4f}"
            row['LogLoss'] = f"{result['metrics']['log_loss']:.4f}"
            row['Strategy'] = result['strategy']
        
        summary_data.append(row)
    
    summary_df = pd.DataFrame(summary_data)
    print(summary_df.to_string(index=False))
    print("="*70)
    
    return summary_df
```

**Итог:** снижение точности может быть связано с более сложной структурой классов либо с менее выраженными различиями между объектами воздействия.

#### Модель 3: Успешность атаки (бинарная классификация)

* **Точность:** 0,5019 ± 0,0239 — практически на уровне случайного угадывания (50 %).
* **Классы:** 0 и 1 — простая бинарная задача.
* **Причина низкой точности:** признаки, важные для других моделей («Идентификатор УБИ», «Код предприятия»), здесь не дают достаточной разделительной способности. Возможно, успех атаки зависит от скрытых факторов или сложных взаимодействий признаков, которые дерево решений не улавливает.
* **Ключевые признаки:** «Код предприятия», «Количество хостов», «Регион размещения» и временные признаки («hour», «day»).

**Итог:** текущая модель бесполезна для практического применения. Требуется:
* добавить признаки, напрямую отражающие условия проведения атаки;
* попробовать другие алгоритмы (логистическая регрессия, SVM, ансамбли);
* проверить баланс классов (возможно, один класс сильно преобладает).

#### Модель 4: Период атаки (Moving Average Baseline)

* **Точность:** максимум 0,2950 (при окне 5), остальные конфигурации — около 0,2425.
* **Классы:** 4 периода суток («Night», «Morning», «Afternoon», «Evening»).
* **Наблюдения:**
    * стратегия (`mode`, `ema`) почти не влияет на результат;
    * размер окна критически важен: окно 5 даёт лучший результат, большие окна (20–100) не улучшают точность;
    * низкая точность говорит о том, что временные паттерны атак либо отсутствуют, либо слишком зашумлены.
* **Эвристическая важность признаков:** автокорреляция (0,5), историческая частота (0,3), цикличность (0,2) — гипотеза о наличии временных закономерностей не подтверждается данными.

**Итог:** простой временной baseline плохо предсказывает период атаки

1. **Многоклассовые задачи** (Модели 1 и 2) работают лучше, чем бинарные или временные (Модели 3 и 4), но есть дисбаланс классов.

    * применить техники балансировки;
    * объединить редкие классы в супер‑категории.

    * протестировать другие алгоритмы;
    * проверить баланс классов.
5. **Низкая точность Модели 4** указывает на отсутствие явных временных паттернов. Рекомендации:
    * изучить распределение атак по периодам (гистограмма);
    * включить внешние факторы (праздники, смены, пиковые нагрузки);
    * попробовать сложные временные модели.
