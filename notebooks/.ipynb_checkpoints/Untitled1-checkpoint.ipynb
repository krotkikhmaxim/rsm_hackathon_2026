```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from pathlib import Path
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
import seaborn as sns

from catboost import CatBoostClassifier, CatBoostRegressor, Pool

from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report,
    mean_absolute_error, mean_squared_error, r2_score
)

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["font.size"] = 12

print("✅ Все библиотеки импортированы")
```

```python
def load_data():
    """
    Загрузка и объединение данных инцидентов и справочника угроз.
    Адаптируйте пути к вашим файлам.
    """
    base_path = Path("..")
    inc_path = (base_path / "data/raw/incidents_2000.csv").resolve()
    thr_path = (base_path / "data/raw/thrlist.csv").resolve()

    print(f"📂 Загрузка инцидентов: {inc_path}")
    print(f"📂 Загрузка справочника угроз: {thr_path}")

    inc = pd.read_csv(inc_path)
    thr = pd.read_csv(thr_path, header=1)

    df = inc.merge(
        thr,
        how="left",
        left_on="Код реализованной угрозы",
        right_on="Идентификатор УБИ"
    )

    print(f"✅ Загружено {len(df)} записей, {df.shape[1]} колонок")
    return df


def convert_dtypes(df: pd.DataFrame) -> pd.DataFrame:
    """Приведение типов данных."""
    df = df.copy()

    # Даты
    date_cols = [
        "Региональное время",
        "Дата инцидента",
        "Дата включения угрозы в БнД УБИ",
        "Дата последнего изменения данных"
    ]
    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], dayfirst=True, errors="coerce")

    # Числовые
    num_cols = [
        "Количество хостов", "Успех",
        "Код реализованной угрозы", "Код предприятия", "Идентификатор УБИ"
    ]
    for col in num_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    return df


def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """Базовая очистка данных."""
    df = df.copy()

    # Удаление дубликатов по предприятию (оставляем последний инцидент)
    df = (df
          .sort_values('Региональное время', ascending=False)
          .drop_duplicates(subset=['Код предприятия'], keep='first'))

    # Удаление архивных угроз
    if "Статус угрозы" in df.columns:
        df = df[df["Статус угрозы"] != "Архивная"]

    # Заполнение пропусков в источнике угрозы
    col_src = "Источник угрозы (характеристика и потенциал нарушителя)"
    if col_src in df.columns:
        mode_val = df[col_src].mode()
        if len(mode_val) > 0:
            df[col_src] = df[col_src].fillna(mode_val.iloc[0])

    return df.reset_index(drop=True)


# === Загрузка ===
df_raw = load_data()
df_raw = convert_dtypes(df_raw)
df_raw = clean_data(df_raw)

print(f"\n📊 Итоговый датасет: {df_raw.shape[0]} строк, {df_raw.shape[1]} колонок")
df_raw.head(3)
```

```python
# Основные колонки для задачи
FEATURE_COLS = ['Тип предприятия', 'Код предприятия', 'Количество хостов', 'Регион размещения предприятия']
TARGET_CLF = 'Код реализованной угрозы'
TARGET_REG = 'Региональное время'

print("=" * 60)
print("ИНФОРМАЦИЯ О ДАННЫХ")
print("=" * 60)

# Проверяем наличие всех нужных колонок
required_cols = FEATURE_COLS + [TARGET_CLF, TARGET_REG]
missing_cols = [c for c in required_cols if c not in df_raw.columns]
if missing_cols:
    print(f"⚠️ Отсутствуют колонки: {missing_cols}")
else:
    print("✅ Все необходимые колонки присутствуют")

print(f"\nРазмер датасета: {df_raw.shape}")
print(f"\nПропуски в ключевых колонках:")
for col in required_cols:
    if col in df_raw.columns:
        n_miss = df_raw[col].isna().sum()
        print(f"  {col}: {n_miss} ({n_miss/len(df_raw)*100:.1f}%)")

print(f"\n📌 Уникальных кодов угроз: {df_raw[TARGET_CLF].nunique()}")
print(f"📌 Диапазон времени: {df_raw[TARGET_REG].min()} — {df_raw[TARGET_REG].max()}")
print(f"\n📌 Типы предприятий: {df_raw['Тип предприятия'].nunique()}")
print(f"📌 Регионов: {df_raw['Регион размещения предприятия'].nunique()}")

print("\nТоп-10 кодов угроз:")
display(df_raw[TARGET_CLF].value_counts().head(10))
```

```python
# === Подготовка рабочего DataFrame ===
df = df_raw.copy()

# Удаляем строки с пропусками в ключевых полях
df = df.dropna(subset=FEATURE_COLS + [TARGET_CLF, TARGET_REG]).copy()
print(f"После удаления пропусков: {len(df)} строк")

# Сортировка по времени (критично для временной валидации!)
df = df.sort_values(TARGET_REG).reset_index(drop=True)

# === Признаки ===
X = df[FEATURE_COLS].copy()

# === Целевая 1: Код угрозы (классификация) ===
y_threat = df[TARGET_CLF].astype(int).copy()

# === Целевая 2: Региональное время → Unix timestamp (секунды) ===
try:
    y_time = df[TARGET_REG].astype(np.int64) // 10**9  # наносекунды → секунды
    print(f"✅ Время преобразовано в Unix timestamp")
    print(f"   Диапазон: {y_time.min()} — {y_time.max()}")
    print(f"   Что соответствует: {pd.Timestamp(y_time.min(), unit='s')} — {pd.Timestamp(y_time.max(), unit='s')}")
except Exception as e:
    print(f"❌ Ошибка преобразования времени: {e}")
    # Fallback: попробуем парсить заново
    df[TARGET_REG] = pd.to_datetime(df[TARGET_REG], errors='coerce')
    df = df.dropna(subset=[TARGET_REG])
    y_time = df[TARGET_REG].astype(np.int64) // 10**9
    X = df[FEATURE_COLS].copy()
    y_threat = df[TARGET_CLF].astype(int).copy()

# === Категориальные признаки для CatBoost ===
cat_features = ['Тип предприятия', 'Код предприятия', 'Регион размещения предприятия']
# Приводим к строкам (CatBoost требует str для cat_features)
for col in cat_features:
    X[col] = X[col].astype(str)

print(f"\n📐 Размер X: {X.shape}")
print(f"📐 Классов угроз: {y_threat.nunique()}")
print(f"📐 Категориальные признаки: {cat_features}")
print(f"\nПервые строки X:")
display(X.head())
```

```python
# === Temporal Hold-out Split ===
split_idx = int(len(X) * 0.8)

X_train = X.iloc[:split_idx].copy()
X_test = X.iloc[split_idx:].copy()

y_train_threat = y_threat.iloc[:split_idx].copy()
y_test_threat = y_threat.iloc[split_idx:].copy()

y_train_time = y_time.iloc[:split_idx].copy()
y_test_time = y_time.iloc[split_idx:].copy()

# Сохраняем оригинальные даты для интерпретации
dates_train = df[TARGET_REG].iloc[:split_idx].copy()
dates_test = df[TARGET_REG].iloc[split_idx:].copy()

print("=" * 60)
print("РАЗДЕЛЕНИЕ ДАННЫХ (Temporal Hold-out)")
print("=" * 60)
print(f"Train: {len(X_train)} ({len(X_train)/len(X)*100:.0f}%)")
print(f"Test:  {len(X_test)} ({len(X_test)/len(X)*100:.0f}%)")
print(f"\nTrain период: {dates_train.min()} — {dates_train.max()}")
print(f"Test период:  {dates_test.min()} — {dates_test.max()}")
print(f"\nКлассов угроз в train: {y_train_threat.nunique()}")
print(f"Классов угроз в test:  {y_test_threat.nunique()}")

# Проверка: есть ли в test классы, которых нет в train?
unseen_classes = set(y_test_threat.unique()) - set(y_train_threat.unique())
if unseen_classes:
    print(f"\n⚠️ В test есть {len(unseen_classes)} классов, отсутствующих в train: {unseen_classes}")
else:
    print(f"\n✅ Все классы test присутствуют в train")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Train
threat_counts_train = y_train_threat.value_counts().head(20)
axes[0].barh(threat_counts_train.index.astype(str), threat_counts_train.values, color='steelblue')
axes[0].set_title(f'Train: Топ-20 кодов угроз (всего {y_train_threat.nunique()} классов)', fontsize=13)
axes[0].set_xlabel('Количество')
axes[0].set_ylabel('Код угрозы')
axes[0].invert_yaxis()

# Test
threat_counts_test = y_test_threat.value_counts().head(20)
axes[1].barh(threat_counts_test.index.astype(str), threat_counts_test.values, color='coral')
axes[1].set_title(f'Test: Топ-20 кодов угроз (всего {y_test_threat.nunique()} классов)', fontsize=13)
axes[1].set_xlabel('Количество')
axes[1].set_ylabel('Код угрозы')
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig('threat_distribution_train_test.png', dpi=150, bbox_inches='tight')
plt.show()

print("📊 Распределение сильно несбалансировано — это типично для задач ИБ")
```

```python
def temporal_cross_validation(X_train, y_train_threat, y_train_time, cat_features, n_splits=5):
    """
    Временная кросс-валидация для обеих моделей одновременно.
    """
    tscv = TimeSeriesSplit(n_splits=n_splits)
    
    clf_results = []
    reg_results = []
    
    for fold, (tr_idx, val_idx) in enumerate(tscv.split(X_train), start=1):
        print(f"\n{'='*40} Fold {fold}/{n_splits} {'='*40}")
        
        X_tr = X_train.iloc[tr_idx]
        X_val = X_train.iloc[val_idx]
        
        y_tr_th = y_train_threat.iloc[tr_idx]
        y_val_th = y_train_threat.iloc[val_idx]
        
        y_tr_tm = y_train_time.iloc[tr_idx]
        y_val_tm = y_train_time.iloc[val_idx]
        
        print(f"  Train: {len(X_tr)}, Val: {len(X_val)}")
        
        # --- Классификатор ---
        clf = CatBoostClassifier(
            cat_features=cat_features,
            loss_function='MultiClass',
            eval_metric='MultiClass',
            iterations=500,
            depth=6,
            learning_rate=0.05,
            verbose=0,
            random_seed=42,
            auto_class_weights='Balanced'
        )
        clf.fit(X_tr, y_tr_th, eval_set=(X_val, y_val_th), early_stopping_rounds=50)
        
        y_pred_th = clf.predict(X_val).flatten()
        # CatBoost predict может вернуть строки — приводим к int
        y_pred_th = pd.Series(y_pred_th).astype(int).values
        
        clf_metrics = {
            'fold': fold,
            'accuracy': accuracy_score(y_val_th, y_pred_th),
            'f1_macro': f1_score(y_val_th, y_pred_th, average='macro', zero_division=0),
            'f1_weighted': f1_score(y_val_th, y_pred_th, average='weighted', zero_division=0),
            'precision_macro': precision_score(y_val_th, y_pred_th, average='macro', zero_division=0),
            'recall_macro': recall_score(y_val_th, y_pred_th, average='macro', zero_division=0),
        }
        clf_results.append(clf_metrics)
        
        # --- Регрессор ---
        reg = CatBoostRegressor(
            cat_features=cat_features,
            loss_function='RMSE',
            iterations=500,
            depth=6,
            learning_rate=0.05,
            verbose=0,
            random_seed=42
        )
        reg.fit(X_tr, y_tr_tm, eval_set=(X_val, y_val_tm), early_stopping_rounds=50)
        
        y_pred_tm = reg.predict(X_val)
        
        mae_sec = mean_absolute_error(y_val_tm, y_pred_tm)
        rmse_sec = np.sqrt(mean_squared_error(y_val_tm, y_pred_tm))
        r2 = r2_score(y_val_tm, y_pred_tm)
        
        reg_metrics = {
            'fold': fold,
            'MAE_sec': mae_sec,
            'MAE_days': mae_sec / 86400,
            'RMSE_sec': rmse_sec,
            'RMSE_days': rmse_sec / 86400,
            'R2': r2
        }
        reg_results.append(reg_metrics)
        
        print(f"  CLF: accuracy={clf_metrics['accuracy']:.4f}, f1_macro={clf_metrics['f1_macro']:.4f}")
        print(f"  REG: MAE={mae_sec/86400:.2f} дней, RMSE={rmse_sec/86400:.2f} дней, R²={r2:.4f}")
    
    return pd.DataFrame(clf_results), pd.DataFrame(reg_results)


# === Запуск кросс-валидации ===
print("🔄 Запуск временной кросс-валидации (5 фолдов)...")
cv_clf_results, cv_reg_results = temporal_cross_validation(
    X_train, y_train_threat, y_train_time, cat_features, n_splits=5
)
```

```python
print("\n" + "=" * 60)
print("РЕЗУЛЬТАТЫ КРОСС-ВАЛИДАЦИИ: КЛАССИФИКАЦИЯ (Код угрозы)")
print("=" * 60)
display(cv_clf_results)
print("\nСреднее по фолдам:")
display(cv_clf_results.drop(columns=['fold']).mean().to_frame('mean').T)
print("\nСтд по фолдам:")
display(cv_clf_results.drop(columns=['fold']).std().to_frame('std').T)

print("\n" + "=" * 60)
print("РЕЗУЛЬТАТЫ КРОСС-ВАЛИДАЦИИ: РЕГРЕССИЯ (Время)")
print("=" * 60)
display(cv_reg_results)
print("\nСреднее по фолдам:")
display(cv_reg_results.drop(columns=['fold']).mean().to_frame('mean').T)
```

```python
print("=" * 60)
print("ОБУЧЕНИЕ МОДЕЛИ 1: CatBoostClassifier (Код угрозы)")
print("=" * 60)

clf_final = CatBoostClassifier(
    cat_features=cat_features,
    loss_function='MultiClass',
    eval_metric='MultiClass',
    iterations=1000,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=3,
    verbose=100,
    random_seed=42,
    auto_class_weights='Balanced',
    early_stopping_rounds=100
)

clf_final.fit(
    X_train, y_train_threat,
    eval_set=(X_test, y_test_threat),
    plot=False
)

print(f"\n✅ Классификатор обучен. Лучшая итерация: {clf_final.get_best_iteration()}")

print("\n" + "=" * 60)
print("ОБУЧЕНИЕ МОДЕЛИ 2: CatBoostRegressor (Время)")
print("=" * 60)

reg_final = CatBoostRegressor(
    cat_features=cat_features,
    loss_function='RMSE',
    eval_metric='RMSE',
    iterations=1000,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=3,
    verbose=100,
    random_seed=42,
    early_stopping_rounds=100
)

reg_final.fit(
    X_train, y_train_time,
    eval_set=(X_test, y_test_time),
    plot=False
)

print(f"\n✅ Регрессор обучен. Лучшая итерация: {reg_final.get_best_iteration()}")
```

```python
# === Прогноз: Классификация ===
pred_threat = clf_final.predict(X_test).flatten()
pred_threat = pd.Series(pred_threat).astype(int).values

pred_threat_proba = clf_final.predict_proba(X_test)
# Максимальная вероятность для каждого предсказания
pred_threat_max_proba = pred_threat_proba.max(axis=1)

# === Прогноз: Регрессия ===
pred_time_unix = reg_final.predict(X_test)

# Преобразуем обратно в читаемые даты
pred_time_datetime = pd.to_datetime(pred_time_unix, unit='s', errors='coerce')
actual_time_datetime = pd.to_datetime(y_test_time.values, unit='s', errors='coerce')

# Ошибка в днях
time_error_seconds = np.abs(y_test_time.values - pred_time_unix)
time_error_days = time_error_seconds / 86400

print("✅ Прогнозы получены")
print(f"   Предсказано угроз: {len(pred_threat)}")
print(f"   Уникальных предсказанных кодов: {len(np.unique(pred_threat))}")
print(f"   Средняя уверенность классификатора: {pred_threat_max_proba.mean():.2%}")
print(f"   Средняя ошибка времени: {time_error_days.mean():.2f} дней")
```

```python
cm = confusion_matrix(y_test_threat, pred_threat)

# Если классов много, показываем топ-15 по частоте
top_classes = y_test_threat.value_counts().head(15).index.tolist()
mask_test = y_test_threat.isin(top_classes)
mask_pred = pd.Series(pred_threat).isin(top_classes).values
mask = mask_test.values & mask_pred

if mask.sum() > 0 and len(top_classes) > 1:
    cm_top = confusion_matrix(
        y_test_threat.values[mask], 
        pred_threat[mask],
        labels=sorted(top_classes)
    )
    
    fig, ax = plt.subplots(figsize=(14, 12))
    sns.heatmap(
        cm_top, 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=sorted(top_classes),
        yticklabels=sorted(top_classes),
        ax=ax
    )
    ax.set_title('Confusion Matrix (Топ-15 классов угроз)', fontsize=14)
    ax.set_xlabel('Предсказанный код угрозы', fontsize=12)
    ax.set_ylabel('Фактический код угрозы', fontsize=12)
    plt.tight_layout()
    plt.savefig('confusion_matrix_top15.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print("⚠️ Недостаточно данных для построения confusion matrix по топ-классам")
    
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(cm, cmap='Blues', ax=ax)
    ax.set_title('Confusion Matrix (все классы)', fontsize=14)
    ax.set_xlabel('Предсказанный')
    ax.set_ylabel('Фактический')
    plt.tight_layout()
    plt.show()
```

```python
print("=" * 60)
print("МЕТРИКИ РЕГРЕССИИ: Региональное время")
print("=" * 60)

mae_sec = mean_absolute_error(y_test_time, pred_time_unix)
rmse_sec = np.sqrt(mean_squared_error(y_test_time, pred_time_unix))
r2 = r2_score(y_test_time, pred_time_unix)

mae_days = mae_sec / 86400
rmse_days = rmse_sec / 86400
mae_hours = mae_sec / 3600
rmse_hours = rmse_sec / 3600

metrics_reg = {
    'MAE (секунды)': mae_sec,
    'MAE (часы)': mae_hours,
    'MAE (дни)': mae_days,
    'RMSE (секунды)': rmse_sec,
    'RMSE (часы)': rmse_hours,
    'RMSE (дни)': rmse_days,
    'R²': r2
}

for name, val in metrics_reg.items():
    print(f"  {name}: {val:.4f}")

print(f"\n📌 Интерпретация:")
print(f"   Средняя ошибка предсказания времени: ~{mae_days:.1f} дней ({mae_hours:.1f} часов)")
print(f"   R² = {r2:.4f} — {'хорошо' if r2 > 0.5 else 'модель объясняет мало дисперсии'}")
```

```python
def predict_threat_and_time(
    enterprise_type: str,
    enterprise_code: int,
    n_hosts: int,
    region: str,
    clf_model=None,
    reg_model=None,
    clf_path: str = "models/catboost_threat_classifier.cbm",
    reg_path: str = "models/catboost_time_regressor.cbm",
    top_k: int = 3
) -> dict:
    """
    Совместный прогноз угрозы и времени для нового предприятия.
    
    Parameters:
    -----------
    enterprise_type : str — Тип предприятия
    enterprise_code : int — Код предприятия
    n_hosts : int — Количество хостов
    region : str — Регион размещения
    clf_model : CatBoostClassifier, optional — загруженная модель (или загрузится из файла)
    reg_model : CatBoostRegressor, optional — загруженная модель (или загрузится из файла)
    top_k : int — сколько наиболее вероятных угроз показать
    
    Returns:
    --------
    dict с результатами прогноза
    """
    # Загрузка моделей если не переданы
    if clf_model is None:
        clf_model = CatBoostClassifier()
        clf_model.load_model(clf_path)
    
    if reg_model is None:
        reg_model = CatBoostRegressor()
        reg_model.load_model(reg_path)
    
    # Формируем входные данные
    input_df = pd.DataFrame({
        'Тип предприятия': [str(enterprise_type)],
        'Код предприятия': [str(enterprise_code)],
        'Количество хостов': [n_hosts],
        'Регион размещения предприятия': [str(region)]
    })
    
    # Прогноз угрозы
    threat_pred = int(clf_model.predict(input_df).flatten()[0])
    threat_proba = clf_model.predict_proba(input_df)[0]
    threat_classes = clf_model.classes_
    
    # Топ-K наиболее вероятных угроз
    top_indices = np.argsort(threat_proba)[::-1][:top_k]
    top_threats = [
        {
            'code': int(threat_classes[i]),
            'probability': float(threat_proba[i]),
            'probability_pct': f"{threat_proba[i]*100:.1f}%"
        }
        for i in top_indices
    ]
    
    # Прогноз времени
    time_pred_unix = float(reg_model.predict(input_df)[0])
    time_pred_dt = pd.Timestamp(time_pred_unix, unit='s')
    
    # Через сколько дней от текущего момента
    now = pd.Timestamp.now()
    days_from_now = (time_pred_dt - now).total_seconds() / 86400
    
    result = {
        'input': {
            'enterprise_type': enterprise_type,
            'enterprise_code': enterprise_code,
            'n_hosts': n_hosts,
            'region': region
        },
        'threat_prediction': {
            'predicted_code': threat_pred,
            'confidence': f"{threat_proba.max()*100:.1f}%",
            'top_threats': top_threats
        },
        'time_prediction': {
            'predicted_datetime': time_pred_dt.strftime('%Y-%m-%d %H:%M:%S'),
            'unix_timestamp': time_pred_unix,
            'days_from_now': round(days_from_now, 1)
        },
        'interpretation': (
            f"Предприятие '{enterprise_type}' (код {enterprise_code}, регион '{region}', "
            f"{n_hosts} хостов) — с вероятностью {threat_proba.max()*100:.0f}% "
            f"ожидается угроза УБИ.{threat_pred:03d}, "
            f"примерное время: {time_pred_dt.strftime('%Y-%m-%d %H:%M')} "
            f"({'через ' + str(abs(round(days_from_now))) + ' дней' if days_from_now > 0 else str(abs(round(days_from_now))) + ' дней назад'})"
        )
    }
    
    return result


# === Пример использования ===
print("=" * 60)
print("🔮 ПРИМЕР ИНФЕРЕНСА ДЛЯ НОВОГО ПРЕДПРИЯТИЯ")
print("=" * 60)

# Берём параметры из реальных данных для демонстрации
sample_row = X_test.iloc[0]

result = predict_threat_and_time(
    enterprise_type=sample_row['Тип предприятия'],
    enterprise_code=int(sample_row['Код предприятия']),
    n_hosts=int(sample_row['Количество хостов']),
    region=sample_row['Регион размещения предприятия'],
    clf_model=clf_final,
    reg_model=reg_final,
    top_k=5
)

print(f"\n📌 Интерпретация:")
print(f"   {result['interpretation']}")

print(f"\n📋 Топ-5 наиболее вероятных угроз:")
for t in result['threat_prediction']['top_threats']:
    print(f"   УБИ.{t['code']:03d} — {t['probability_pct']}")

print(f"\n⏰ Предсказанное время: {result['time_prediction']['predicted_datetime']}")
print(f"   ({result['time_prediction']['days_from_now']} дней от текущего момента)")
```

```python
from sklearn.multioutput import MultiOutputRegressor

print("=" * 60)
print("АЛЬТЕРНАТИВА: MultiOutputRegressor (CatBoost)")
print("=" * 60)

# Формируем двумерный target: [код_угрозы, unix_timestamp]
y_train_multi = np.column_stack([y_train_threat.values, y_train_time.values])
y_test_multi = np.column_stack([y_test_threat.values, y_test_time.values])

# MultiOutputRegressor оборачивает CatBoostRegressor
multi_reg = MultiOutputRegressor(
    CatBoostRegressor(
        cat_features=[0, 1, 3],  # индексы категориальных колонок
        iterations=500,
        depth=6,
        learning_rate=0.05,
        verbose=0,
        random_seed=42
    )
)

print("🔄 Обучение MultiOutputRegressor...")
multi_reg.fit(X_train, y_train_multi)

y_pred_multi = multi_reg.predict(X_test)

# Разделяем предсказания
pred_threat_multi = np.round(y_pred_multi[:, 0]).astype(int)
pred_time_multi = y_pred_multi[:, 1]

# Метрики для угрозы (как регрессия → округление)
acc_multi = accuracy_score(y_test_threat, pred_threat_multi)
f1_multi = f1_score(y_test_threat, pred_threat_multi, average='macro', zero_division=0)

# Метрики для времени
mae_multi = mean_absolute_error(y_test_time, pred_time_multi) / 86400
r2_multi = r2_score(y_test_time, pred_time_multi)

print(f"\n📊 MultiOutput результаты:")
print(f"   Угроза — Accuracy: {acc_multi:.4f}, F1 macro: {f1_multi:.4f}")
print(f"   Время  — MAE: {mae_multi:.2f} дней, R²: {r2_multi:.4f}")

print(f"\n📊 Сравнение с отдельными моделями:")
print(f"   Отдельные: Accuracy={acc:.4f}, F1={f1_macro:.4f}, MAE={mae_days:.2f} дней, R²={r2:.4f}")
print(f"   MultiOut:  Accuracy={acc_multi:.4f}, F1={f1_multi:.4f}, MAE={mae_multi:.2f} дней, R²={r2_multi:.4f}")
print(f"\n💡 Вывод: отдельные модели обычно лучше, т.к. классификация через регрессию теряет дискретность")
```
