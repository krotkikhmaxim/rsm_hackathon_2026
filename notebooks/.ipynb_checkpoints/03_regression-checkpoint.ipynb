# 03_regression — Регрессия: время до следующей атаки

**Цель ноутбука:**  
Обучить регрессионные модели (CatBoostRegressor) для предсказания `days_until_next_attack` —  
количества дней до ближайшей следующей атаки в каждом сегменте (infrastructure × threat).

**Архитектура:**

```
dataset_features.parquet
        ↓
[Для каждого сегмента (infra × threat)]
        ↓
  Временной split (70% train / 15% val / 15% test)
        ↓
  CatBoostRegressor
        ↓
  Оценка: MAE, RMSE, R²
        ↓
  Сохранение моделей + метаданных
```

**Входные файлы:**
- `data/interim/dataset_features.parquet`
- `data/interim/feature_meta.json`

**Выходные файлы:**
- `mvp/models/regression/<infra>_<threat>.cbm`
- `mvp/models/results_regression.csv`

**Связанные ноутбуки:**
- `02_feature_engineering_v2.ipynb` — построение признаков и цели `days_until_next_attack`
- `03_multiclass.ipynb` — мультиклассовая классификация по типам окон

````python
import warnings
warnings.filterwarnings('ignore')

import json
import numpy as np
import pandas as pd
from pathlib import Path

import catboost as cb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', context='notebook')
plt.rcParams['figure.figsize'] = (12, 5)

print("Все библиотеки загружены")
````

## 2. Загрузка данных

````python
base_path = Path('..')
data_dir = base_path / 'data' / 'interim'

dataset_model = pd.read_parquet(data_dir / 'dataset_features.parquet')

with open(data_dir / 'feature_meta.json', 'r') as f:
    meta = json.load(f)

feature_cols = meta['feature_cols']
cat_features = meta['cat_features']

TARGET_COL = 'days_until_next_attack'

# Удаляем строки без цели (конец горизонта наблюдений)
df_reg = dataset_model.dropna(subset=[TARGET_COL]).copy()

print(f"Датасет (до фильтрации): {dataset_model.shape}")
print(f"Датасет (после удаления NaN в target): {df_reg.shape}")
print(f"Признаков: {len(feature_cols)}")
print(f"\nРаспределение {TARGET_COL}:")
print(df_reg[TARGET_COL].describe().round(2))
print(f"\nДоля строк с атакой в день: {(df_reg[TARGET_COL] == 0).mean():.3f}")
````

## 4. Вспомогательные функции

### `add_rolling` — скользящие признаки

Та же функция из `03_model_training.ipynb` — пересчитываем rolling внутри каждого split,
чтобы не допустить утечки данных между train/val/test.


````python
def add_rolling(df_part: pd.DataFrame) -> pd.DataFrame:
    """
    Добавляет скользящие признаки к датафрейму сегмента.
    df_part должен быть отсортирован по дате.
    """
    df_part = df_part.copy()
    grp = df_part.groupby(['infrastructure_cluster', 'threat_cluster'])

    for window, name in [(3, 'inc_3d_sum'), (7, 'inc_7d_sum'), (30, 'inc_30d_sum')]:
        df_part[name] = (
            grp['incidents_count_day']
            .shift(1).rolling(window, min_periods=1).sum()
            .reset_index(drop=True).fillna(0)
        )

    for window, name in [(7, 'succ_7d_sum'), (30, 'succ_30d_sum')]:
        df_part[name] = (
            grp['success_count_day']
            .shift(1).rolling(window, min_periods=1).sum()
            .reset_index(drop=True).fillna(0)
        )

    df_part['had_incident_prev_1d'] = (grp['lag_inc_1d'].transform(lambda x: (x > 0).astype(int))
                                       if 'lag_inc_1d' in df_part.columns
                                       else 0)
    df_part['had_incident_prev_3d'] = (df_part['inc_3d_sum'] > 0).astype(int)
    df_part['had_incident_prev_7d'] = (df_part['inc_7d_sum'] > 0).astype(int)

    return df_part


print("Функция add_rolling определена")
````

## 5. Функция обучения регрессионных моделей по сегментам

### Ключевые отличия от бинарного классификатора:

| Параметр | Классификатор | Регрессор |
|---------|---------------|-----------|
| `loss_function` | `Logloss` | `RMSE` (или `MAE`) |
| Метрики | ROC-AUC, PR-AUC | MAE, RMSE, R² |
| Дисбаланс | `scale_pos_weight` | не требуется |
| Target | {0, 1} | ≥ 0, вещественный |

### Целевое преобразование
Используем `log1p(target)` при обучении для стабилизации дисперсии,  
затем `expm1(pred)` при инференсе — это стандартная практика для time-to-event регрессии.


````python
def train_regression_models_by_segments(
    data: pd.DataFrame,
    feature_cols: list,
    target_col: str,
    cat_features: list,
    min_rows: int = 120,
    min_events: int = 10,
    use_log_target: bool = True,
    verbose: bool = True
) -> tuple:
    """
    Обучает CatBoostRegressor для каждого сегмента (infra × threat).

    Параметры:
    ----------
    data : pd.DataFrame
        Датасет с признаками и целевой переменной.
    feature_cols : list
        Список имён признаков.
    target_col : str
        Имя целевой переменной ('days_until_next_attack').
    cat_features : list
        Список категориальных признаков.
    min_rows : int
        Минимум строк для сегмента (по умолчанию 120).
    min_events : int
        Минимум строк с атакой (target == 0) в обучающей выборке.
    use_log_target : bool
        Применять ли log1p-трансформацию к целевой переменной.
    verbose : bool
        Печатать прогресс.

    Возвращает:
    ----------
    models : dict  {(infra, threat): CatBoostRegressor}
    results_df : pd.DataFrame  с метриками по сегментам
    """
    models = {}
    results = []

    infra_values = sorted(data['infrastructure_cluster'].astype(str).unique())
    threat_values = sorted(data['threat_cluster'].astype(int).unique())
    total = len(infra_values) * len(threat_values)
    done = 0

    for infra in infra_values:
        for threat in threat_values:
            done += 1
            part = data[
                (data['infrastructure_cluster'].astype(str) == str(infra)) &
                (data['threat_cluster'].astype(int) == int(threat)) &
                (data[target_col].notna())
            ].copy()

            if len(part) < min_rows:
                if verbose:
                    print(f"  [{done}/{total}] infra={infra} threat={threat} — пропуск (<{min_rows} строк)")
                continue

            # Временной split 70/15/15
            part = part.sort_values('date').reset_index(drop=True)
            unique_dates = np.array(sorted(part['date'].unique()))
            n = len(unique_dates)

            if n < 30:
                continue

            train_end = int(n * 0.70)
            val_end   = int(n * 0.85)

            train_dates = unique_dates[:train_end]
            val_dates   = unique_dates[train_end:val_end]
            test_dates  = unique_dates[val_end:]

            train_df = add_rolling(part[part['date'].isin(train_dates)].copy())
            val_df   = add_rolling(part[part['date'].isin(val_dates)].copy())
            test_df  = add_rolling(part[part['date'].isin(test_dates)].copy())

            # Убираем NaN в target
            train_df = train_df.dropna(subset=[target_col])
            val_df   = val_df.dropna(subset=[target_col])
            test_df  = test_df.dropna(subset=[target_col])

            if len(train_df) < min_rows or (train_df[target_col] == 0).sum() < min_events:
                continue

            # Подготовка признаков
            local_cats = [c for c in cat_features if c in feature_cols]
            avail_feats = [f for f in feature_cols if f in train_df.columns]

            def prep(df):
                X = df[avail_feats].copy()
                for c in local_cats:
                    if c in X.columns:
                        X[c] = X[c].astype(str)
                return X

            X_train = prep(train_df)
            X_val   = prep(val_df)
            X_test  = prep(test_df)

            y_train = train_df[target_col].values
            y_val   = val_df[target_col].values
            y_test  = test_df[target_col].values

            # log1p трансформация цели
            if use_log_target:
                y_train_fit = np.log1p(y_train)
                y_val_fit   = np.log1p(y_val)
            else:
                y_train_fit = y_train
                y_val_fit   = y_val

            # Пул для CatBoost
            train_pool = cb.Pool(X_train, y_train_fit, cat_features=local_cats)
            val_pool   = cb.Pool(X_val,   y_val_fit,   cat_features=local_cats)

            model = cb.CatBoostRegressor(
                iterations=600,
                learning_rate=0.05,
                depth=6,
                loss_function='RMSE',
                eval_metric='RMSE',
                early_stopping_rounds=60,
                random_seed=42,
                verbose=0,
                allow_writing_files=False
            )

            model.fit(
                train_pool,
                eval_set=val_pool,
                use_best_model=True
            )

            # Предсказания (обратная трансформация)
            def predict(pool_X):
                raw = model.predict(pool_X)
                return np.expm1(raw) if use_log_target else raw

            val_pred  = np.clip(predict(X_val), 0, None)
            test_pred = np.clip(predict(X_test), 0, None)

            def metrics_dict(y_true, y_pred, prefix):
                mae  = mean_absolute_error(y_true, y_pred)
                rmse = np.sqrt(mean_squared_error(y_true, y_pred))
                r2   = r2_score(y_true, y_pred)
                return {
                    f'{prefix}_mae': mae,
                    f'{prefix}_rmse': rmse,
                    f'{prefix}_r2': r2
                }

            m_val  = metrics_dict(y_val,  val_pred,  'val')
            m_test = metrics_dict(y_test, test_pred, 'test')

            row = {
                'infrastructure_cluster': str(infra),
                'threat_cluster': int(threat),
                'n_train': len(train_df),
                'n_val':   len(val_df),
                'n_test':  len(test_df),
                'pos_rate': (part[target_col] == 0).mean(),
                **m_val,
                **m_test
            }
            results.append(row)
            models[(str(infra), int(threat))] = model

            if verbose:
                print(f"  [{done}/{total}] infra={infra} threat={threat} | "
                      f"val MAE={m_val['val_mae']:.2f} | test MAE={m_test['test_mae']:.2f} | "
                      f"test R²={m_test['test_r2']:.3f}")

    results_df = pd.DataFrame(results)
    return models, results_df


print("Функция train_regression_models_by_segments определена")
````

## 6. Обучение регрессионных моделей

Предсказываем: **за сколько дней произойдёт следующая атака?**

- `days_until_next_attack = 0` → атака сегодня
- `days_until_next_attack = 3` → атака через 3 дня

Применяем `log1p` к цели: это стабилизирует дисперсию, улучшает сходимость  
и делает модель менее чувствительной к редким большим паузам.


````python
print("Обучение регрессионных моделей (days_until_next_attack)...")
print(f"Сегментов для обхода: "
      f"{df_reg['infrastructure_cluster'].nunique()} × "
      f"{df_reg['threat_cluster'].nunique()}")

models_reg, results_reg = train_regression_models_by_segments(
    data=df_reg,
    feature_cols=feature_cols,
    target_col=TARGET_COL,
    cat_features=cat_features,
    min_rows=120,
    min_events=10,
    use_log_target=True,
    verbose=True
)

print(f"\n{'='*80}")
print(f"Обучено моделей: {len(models_reg)}")
print(f"\nТаблица результатов (по убыванию test R²):")
display(results_reg.sort_values('test_r2', ascending=False).round(4))
````

## 7. Анализ результатов

### Интерпретация метрик

| Метрика | Значение |
|---------|----------|
| **MAE** | Средняя абсолютная ошибка в днях. MAE=2 → ошибаемся на ±2 дня |
| **RMSE** | Корень из среднеквадратичной ошибки. Чувствителен к выбросам |
| **R²** | Доля объяснённой дисперсии. R²=0.6 → модель объясняет 60% вариации |

Для задачи ИБ **MAE ≤ 3 дня** — хороший результат; MAE ≤ 1 дня — отличный.


````python
if len(results_reg) == 0:
    print("Нет обученных моделей — проверьте данные")
else:
    print("=== Сводная статистика метрик ===\n")
    for m in ['test_mae', 'test_rmse', 'test_r2']:
        s = results_reg[m]
        print(f"{m}: mean={s.mean():.3f}, median={s.median():.3f}, "
              f"min={s.min():.3f}, max={s.max():.3f}")

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # MAE histogram
    axes[0].hist(results_reg['test_mae'].dropna(), bins=10,
                 color='teal', edgecolor='white', alpha=0.85)
    axes[0].axvline(results_reg['test_mae'].mean(), color='red',
                    linestyle='--', label=f"Mean={results_reg['test_mae'].mean():.2f}")
    axes[0].axvline(3, color='green', linestyle=':', label='Target ≤3 дней')
    axes[0].set_title('Распределение Test MAE (дней)')
    axes[0].set_xlabel('MAE, дней')
    axes[0].legend()

    # R² histogram
    axes[1].hist(results_reg['test_r2'].dropna(), bins=10,
                 color='steelblue', edgecolor='white', alpha=0.85)
    axes[1].axvline(results_reg['test_r2'].mean(), color='red',
                    linestyle='--', label=f"Mean={results_reg['test_r2'].mean():.3f}")
    axes[1].axvline(0, color='orange', linestyle=':', label='Random baseline')
    axes[1].set_title('Распределение Test R²')
    axes[1].set_xlabel('R²')
    axes[1].legend()

    # MAE vs R² scatter
    axes[2].scatter(results_reg['test_r2'], results_reg['test_mae'],
                    alpha=0.6, color='coral', s=60)
    axes[2].axvline(0, color='gray', linestyle='--', alpha=0.5)
    axes[2].axhline(3, color='green', linestyle=':', alpha=0.7, label='MAE=3')
    axes[2].set_title('Test R² vs Test MAE по сегментам')
    axes[2].set_xlabel('Test R²')
    axes[2].set_ylabel('Test MAE (дней)')
    axes[2].legend()

    plt.tight_layout()
    plt.show()
````

## 9. Важность признаков

Для регрессии важность признаков интерпретируется так же, как для классификатора:
высокий балл = признак сильно влияет на предсказание числа дней до следующей атаки.


````python
if models_reg:
    sample_key = list(models_reg.keys())[0]
    sample_model = models_reg[sample_key]

    avail_feats = [f for f in feature_cols
                   if f in df_reg.columns][:sample_model.feature_count_]

    importance_df = pd.DataFrame({
        'feature': sample_model.feature_names_,
        'importance': sample_model.get_feature_importance()
    }).sort_values('importance', ascending=False)

    print(f"Важность признаков (сегмент infra={sample_key[0]}, threat={sample_key[1]}):")

    fig, ax = plt.subplots(figsize=(10, max(6, len(importance_df) * 0.32)))
    colors = ['#d62728' if any(kw in f for kw in ['inc_', 'days_since', 'streak', 'lag_days'])
              else '#2ca02c' if any(kw in f for kw in ['attack_rate', 'success', 'avg_inter'])
              else '#1f77b4'
              for f in importance_df['feature']]
    ax.barh(importance_df['feature'], importance_df['importance'], color=colors)
    ax.invert_yaxis()
    ax.set_title(f'Feature Importance — Регрессия (infra={sample_key[0]}, threat={sample_key[1]})')
    ax.set_xlabel('Важность (%)')
    plt.tight_layout()
    plt.show()

    # Легенда
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(color='#d62728', label='Лаги / rolling инцидентов'),
        Patch(color='#2ca02c', label='Новые признаки интенсивности'),
        Patch(color='#1f77b4', label='Прочие признаки'),
    ]
    ax.legend(handles=legend_elements, loc='lower right')
    plt.tight_layout()
    plt.show()

    print("\nТоп-10 признаков:")
    display(importance_df.head(10))
````

## 10. Анализ ошибок: визуализация предсказаний

Строим scatter plot: реальные дни vs предсказанные.
Точки на диагонали — идеальные предсказания.


````python
if models_reg:
    # Берём лучший сегмент по R²
    best_seg_row = results_reg.loc[results_reg['test_r2'].idxmax()]
    infra_v = str(best_seg_row['infrastructure_cluster'])
    threat_v = int(best_seg_row['threat_cluster'])

    part = df_reg[
        (df_reg['infrastructure_cluster'].astype(str) == infra_v) &
        (df_reg['threat_cluster'].astype(int) == threat_v)
    ].copy().sort_values('date').reset_index(drop=True)

    unique_dates = np.array(sorted(part['date'].unique()))
    n = len(unique_dates)
    val_end = int(n * 0.85)
    test_dates = unique_dates[val_end:]

    test_df = add_rolling(part[part['date'].isin(test_dates)].copy())
    test_df = test_df.dropna(subset=[TARGET_COL])

    model = models_reg[(infra_v, threat_v)]
    avail_feats = [f for f in feature_cols if f in test_df.columns]
    X_test = test_df[avail_feats].copy()
    local_cats = [c for c in cat_features if c in X_test.columns]
    for c in local_cats:
        X_test[c] = X_test[c].astype(str)

    raw_pred = model.predict(X_test)
    y_pred = np.clip(np.expm1(raw_pred), 0, None)
    y_true = test_df[TARGET_COL].values

    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # Scatter: true vs predicted
    axes[0].scatter(y_true, y_pred, alpha=0.3, s=15, color='steelblue')
    m = max(y_true.max(), y_pred.max())
    axes[0].plot([0, m], [0, m], 'r--', lw=1.5, label='Идеальный прогноз')
    axes[0].set_xlabel('Реальные дни')
    axes[0].set_ylabel('Предсказанные дни')
    axes[0].set_title(f'True vs Predicted\ninfra={infra_v}, threat={threat_v}')
    axes[0].legend()

    # Остатки
    residuals = y_pred - y_true
    axes[1].hist(residuals.clip(-20, 20), bins=40, color='coral', edgecolor='white')
    axes[1].axvline(0, color='black', lw=1.5)
    axes[1].set_title(f'Остатки (pred - true)\nMAE={mae:.2f}, RMSE={rmse:.2f}, R²={r2:.3f}')
    axes[1].set_xlabel('Ошибка (дней)')

    # Временной ряд
    axes[2].plot(test_df['date'].values, y_true, label='Реальные дни', alpha=0.7, lw=1.5)
    axes[2].plot(test_df['date'].values, y_pred, label='Прогноз', alpha=0.7, lw=1.5, linestyle='--')
    axes[2].set_title(f'Временной ряд: прогноз vs факт')
    axes[2].set_xlabel('Дата')
    axes[2].set_ylabel('Дней до атаки')
    axes[2].legend()

    plt.tight_layout()
    plt.show()

    print(f"\nЛучший сегмент: infra={infra_v}, threat={threat_v}")
    print(f"MAE={mae:.2f} дней | RMSE={rmse:.2f} | R²={r2:.3f}")
````

## 12. Утилита инференса

````python
def predict_days_until_attack(
    infra_cluster: int,
    threat_cluster: int,
    feature_values: dict,
    models_dir: Path = Path('mvp/models')
) -> dict:
    """
    Предсказать количество дней до следующей атаки.

    Параметры:
    ----------
    infra_cluster : int     — кластер инфраструктуры (1–4)
    threat_cluster : int    — кластер угрозы (1–6)
    feature_values : dict   — значения признаков
    models_dir : Path       — директория с моделями

    Возвращает:
    ----------
    dict с предсказанием и доверительным интервалом
    """
    model_path = models_dir / 'regression' / f"{infra_cluster}_{threat_cluster}.cbm"

    if not model_path.exists():
        return {'error': f'Модель не найдена: {model_path}'}

    model = cb.CatBoostRegressor()
    model.load_model(str(model_path))

    X = pd.DataFrame([feature_values])
    for col in cat_features:
        if col in X.columns:
            X[col] = X[col].astype(str)

    raw = model.predict(X)[0]
    days = max(0, float(np.expm1(raw)))

    # Грубая оценка интервала (±1 MAE)
    seg_metrics = results_reg[
        (results_reg['infrastructure_cluster'].astype(str) == str(infra_cluster)) &
        (results_reg['threat_cluster'].astype(int) == int(threat_cluster))
    ]
    mae = float(seg_metrics['test_mae'].iloc[0]) if len(seg_metrics) > 0 else None

    return {
        'infrastructure_cluster': infra_cluster,
        'threat_cluster': threat_cluster,
        'predicted_days_until_attack': round(days, 1),
        'model_mae_days': round(mae, 2) if mae else None,
        'low_bound': max(0, round(days - (mae or 0), 1)),
        'high_bound': round(days + (mae or 0), 1)
    }

# Пример использования
if models_reg:
    key = list(models_reg.keys())[0]
    example_row = df_reg[
        (df_reg['infrastructure_cluster'].astype(str) == key[0]) &
        (df_reg['threat_cluster'].astype(int) == key[1])
    ].iloc[-5]

    feat_dict = {f: example_row[f] for f in feature_cols if f in example_row.index}
    result = predict_days_until_attack(key[0], key[1], feat_dict)
    print("Пример предсказания:")
    for k, v in result.items():
        print(f"  {k}: {v}")
````
